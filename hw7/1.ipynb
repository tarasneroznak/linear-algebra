```python
import numpy as np
import matplotlib.pyplot as plt
```

Find the transformation to the new reference frame and the coordinates of the vector a in its basis represented by vectors

u = (0, -0.01) and v = (0.01, 0),

if the origin of the new reference frame is displaced as shown in the picture. 

![alt text](./1_assets/1.png)

```python
t = np.array([4, 5])
a = np.array([-1, 3])

# basis
v = np.array([0.01, 0])
u = np.array([0, -0.01])

S = np.array([v, u]).T
S1 = np.linalg.inv(S)
S1t = S1.dot(t)

M = np.array([
    [S1[0][0], S1[0][1], S1t[0]],
    [S1[1][0], S1[1][1], S1t[1]]
])

a1 = M.dot([a[0], a[1], 1])
a1
```

```python
plt.scatter(*u, c='r')
plt.scatter(*v, c='r')
plt.scatter(*a, c='blue')

plt.scatter(*u1, c='green')
plt.scatter(*v1, c='green')
plt.scatter(*a1, c='yellow')

plt.axis('equal')
plt.show()
```
