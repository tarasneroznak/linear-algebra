```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

"""
Find the vector C, if BC is three times
shorter than AB.
A = (-1, 5, 1)
B = (2, 2, -2)
"""

a = np.array([-1, 5, 1])
b = np.array([2, 2, -2])

x = np.array([-1, 2])
y = np.array([5,2])
z = np.array([1,-2])

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

ax.scatter(x,y,z)

plt.show()
```
