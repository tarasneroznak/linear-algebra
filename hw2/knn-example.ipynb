```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# latent function
def latent_fun(x):
    x = (x - 5) 
    return 10 / (1 + np.exp(-x))
```

```python
Xs = np.linspace(0, 10, 100)
Ys = latent_fun(Xs)
```

```python
plt.figure(figsize=(6,6))
plt.style.use('dark_background')
plt.axis('equal')
plt.plot(Xs, Ys)
plt.show()
```

```python
data = [ (x, latent_fun(x) + 0.2 * np.random.normal()) for x in np.random.uniform(0, 10, size=20)]
```

```python
plt.figure(figsize=(6,6))
plt.style.use('dark_background')
plt.axis('equal')
Xd, Yd = zip(*data)
plt.scatter(Xd, Yd)
plt.plot(Xs, Ys)
plt.show()
```

```python
X1 = np.linspace(0,10,100)
Y1 = []
K = 3
for x1 in X1:
    nearest = sorted([ (np.abs(x-x1), y) for x,y in zip(Xd, Yd) ])
    y1 = sum( y for _,y in nearest[:K]) / K
    Y1.append(y1)

plt.figure(figsize=(6,6))
plt.style.use('dark_background')
plt.axis('equal')
Xd, Yd = zip(*data)
plt.scatter(Xd, Yd)
plt.plot(X1, Y1, 'r')
plt.show()    
```

```python
X2 = np.linspace(0,10,100)
Y2 = []
K = 7
for x2 in X2:
    nearest = sorted([ (np.abs(x-x2), y) for x,y in zip(Xd, Yd) ])
    y2 = sum( y/(dx) for dx,y in nearest[:K]) / sum(1/(dx) for dx,_ in nearest[:K])
    Y2.append(y2)

plt.figure(figsize=(6,6))
plt.style.use('dark_background')
plt.axis('equal')
Xd, Yd = zip(*data)
plt.scatter(Xd, Yd)
plt.plot(X2, Y2, 'r')
plt.show()  
```
