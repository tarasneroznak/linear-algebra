```python
import numpy as np
from matplotlib import pyplot as plt
```

For x from x1=0 to x2=8 plot the function y using KNN. 

Take k=2 and the dataset of pairs (x, y).

data = [ (0, 1), (3, 5), (4, 3), (7, 4) ]

```python
k = 2
x1 = 0 
x2 = 8
data = [ (0, 1), (3, 5), (4, 3), (7, 4) ]

Xd, Yd = zip(*data)
Xq = np.linspace(x1, x2, 50)
Yq = []

for xq in Xq:
    distances = []
    for x,y in data: distances.append((np.abs(x - xq), y))
    sorted_distances = sorted(distances)
    k_sorted_distances = sorted_distances[:k]
    _, ys = zip(*k_sorted_distances)
    yq = sum(ys) / k
    Yq.append(yq)

plt.scatter(Xd, Yd)
plt.plot(Xd, Yd)
plt.scatter(Xq, Yq, s=10, c='r')
plt.plot(Xq, Yq, 'r')
plt.show()
```
