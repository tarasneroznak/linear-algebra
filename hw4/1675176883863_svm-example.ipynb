```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.svm import LinearSVC
```

```python
X = np.random.uniform(-1, 1, size=(50, 2))
eps = 0.1
y = []
for x1, x2 in X:
    if 2 * x1 > x2:
        if np.random.random() > eps:
            y.append(1)
        else:
            y.append(0)
    else:
        if np.random.random() > eps:
            y.append(0)
        else:
            y.append(1)
```

```python
X[:5], y[:5]
```

```python
plt.figure(figsize=(9, 9))
plt.style.use('dark_background')
plt.axis('equal')
plt.plot([-0.5, 0.5], [-1, 1], 'w--')
for (x1, x2), yi in zip(X, y):
    plt.scatter(x1, x2, c="cyan" if yi==0 else "yellow")
plt.show()
```

```python
clf = LinearSVC(C=10, tol=1e-5)
clf.fit(X, y)
d = -clf.intercept_[0]
n = clf.coef_[0]
print(f"d = {d}")
print(f"n = {n}")
```

```python
plt.figure(figsize=(9, 9))
plt.style.use('dark_background')
plt.axis('equal')
plt.plot([-0.5, 0.5], [-1, 1], 'w--')
plt.plot([-0.3, 0.28], [(d + 0.3 * n[0])/n[1], (d - 0.28 * n[0])/n[1]], 'r--')
plt.plot([0, n[0]/4], [0, n[1]/4], "red", linewidth=2)
for (x1, x2), yi in zip(X, y):
    plt.scatter(x1, x2, c="cyan" if yi==0 else "yellow")
plt.show()
```

```python
clf.score(X, y)
```

```python
pars = clf.get_params()
for par, val in pars.items():
    s = " "*(20 - len(par))
    print(f"{par}: {s} {val}")
```
