A plane is given by the equation: x + 2y + 3z = 4. Does it intersect the sphere
with the radius r=1 and the center at [-3, 2, 2]?

```python
r = 1
c = np.array([-3,2,2])
# x + 2y + 3z = 4
normal = np.array([1,2,3])
co = np.array([1,0,1])
# cx = c + (c - co, n) / ||n||^2
cx = c + (np.dot(c - co, normal) / np.linalg.norm(normal) ** 2) 
d = np.linalg.norm(cx - c)

if r > d:
    print('intersect')
else: 
    print('does not intersect')
d
```

```python
# Ax + By + Cz + D = 0
# x + 2y + 3z = 4
# x + 2y + 3z - 4 = 0
# [A,B,C] = [1,2,3] => n - normal vector
# d = |(n * c) + D| / ||n||
r = 1
c = np.array([-3,2,2])
D = -4
n = np.array([1,2,3])
d = np.abs(np.dot(n,c) + D) / np.linalg.norm(n)
if r > d:
    print('intersect')
else: 
    print('does not intersect')
d
```

```python
center = [-3,2,2]
radius = 1

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

u, v = np.mgrid[0:2 * np.pi:50j, 0:np.pi:50j]
x = center[0] + (radius * np.cos(u) * np.sin(v))
y = center[1] + (radius * np.sin(u) * np.sin(v))
z = center[2] + (radius * np.cos(v))

ax.scatter(center[0], center[1], center[2], color='r', alpha=1)
ax.plot_surface(x, y, z, color='b', alpha=0.2)

# =====

# z = (4 - x - 2y) / 3
z = lambda x,y: (4 - x - 2 * y) / 3

tmp = np.linspace(-5,5,30)
x,y = np.meshgrid(tmp,tmp)

ax.plot_surface(x, y, z(x,y))

ax.view_init(0, 153)

ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
