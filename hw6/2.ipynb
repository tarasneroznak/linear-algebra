```python
import numpy as np
import matplotlib.pyplot as plt
```

Find a transformation (a rotation around the origin) that turns the polygon as shown in the graph if the angle is:

a) -45 (clockwise)

b) 30 (counterclockwise)

```python
def rotate(vector, angle):
    return np.dot(vector, np.array([[np.cos(angle), -np.sin(angle)],[np.sin(angle), np.cos(angle)]]))
```

```python
a1 = np.deg2rad(-45)
a2 = np.deg2rad(30)

c = np.array([3, 3])
points = np.array([
    np.array([2, 1]),
    np.array([1, 3]),
    np.array([3, 4]),
    np.array([2, 5]),
    np.array([5, 5]),
    np.array([4, 1]),
    np.array([2, 1]),
])

x, y = zip(*points)
plt.plot(x, y, c='yellow')
plt.scatter(c[0], c[1], c='blue')

x1, y1 = zip(*[rotate(p, a1) for p in points])
c1 = rotate(c, a1)
plt.plot(x1, y1, c='red')
plt.scatter(c1[0], c1[1], c='grey')

x2, y2 = zip(*[rotate(p, a2) for p in points])
c2 = rotate(c, a2)
plt.plot(x2, y2, c='green')
plt.scatter(c2[0], c2[1], c='pink')

plt.axis('equal')
plt.show()
```
