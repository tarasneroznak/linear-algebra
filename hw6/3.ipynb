```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

Find a transformation that rotates the polygon around vector c if:

a) c = [2, 3], angle = -90

b) c = [-1, 4], angle = 45

```python
def rotate(center, vector, angle):
    vector = vector - center
    vector = np.dot(
        vector,
        np.array(
            [[np.cos(angle), np.sin(angle)],
             [-np.sin(angle), np.cos(angle)]]
        ))
    vector = vector + center
    return vector
```

```python
c = np.array([2, 3])
angle = np.deg2rad(-90)
points = np.array([
    np.array([1, 2]),
    np.array([1, 4]),
    np.array([3, 4]),
    np.array([3, 2]),
    np.array([1, 2]),
])

x, y = zip(*points)
plt.plot(x, y, c='blue')
plt.scatter(*c, c='blue')

r = rotate(c, points, angle)
x1, y1 = zip(*r)
plt.plot(x1, y1, c='red')

angle = np.deg2rad(45)

x1,y1 = zip(*rotate(c, points, angle))
plt.plot(x1, y1, c='green')

plt.axis('equal')
plt.show()
```

```python
points = np.array([
    np.array([-2, 3]),
    np.array([-1,5]),
    np.array([0,3]),
    np.array([-2, 3]),
])

c = np.array([-1,4])
x, y = zip(*points)
plt.plot(x, y, c='blue')
plt.scatter(*c, c='blue')

angle = np.deg2rad(-90)

r = rotate(c, points, angle)
x1, y1 = zip(*r)
plt.plot(x1, y1, c='red')

angle = np.deg2rad(45)

x1,y1 = zip(*rotate(c, points, angle))
plt.plot(x1, y1, c='green')

plt.axis('equal')
plt.show()
```
