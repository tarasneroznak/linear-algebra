Find the rank of the given matrices. Are these matrices invertible? Why?

C = [[2 1 3][6 3 9][4 2 6]]

```python
A = np.array([[1, 2],[3, 5]])
B = np.array([[1, 3, -2],[3, 0, 6]])
C = np.array([[2, 1, 3],[6, 3, 9],[4, 2, 6]])

# A(2x2)
# A(3x2)
# A(3x3)

rankA = np.linalg.matrix_rank(A)
detA = int(np.linalg.det(A))
invA = np.linalg.inv(A)
rankB = np.linalg.matrix_rank(B)
# detB = np.linalg.det(B)
invB = np.linalg.pinv(B)
rankC = np.linalg.matrix_rank(C)
detC = int(np.linalg.det(C))
invC = np.linalg.inv(C)

print(C)
print(invC)
print(C.dot(invC))
print(invC.dot(C))
print('======')

print(f"A - rank: {rankA}, det: {detA}, inv: {invA}")
print(f"B - rank: {rankB}, det: {False}, inv: {invB}")
print(f"C - rank: {rankC}, det: {detC}, , inv: {invC}")
```

A rank is 2 and matrix A(2x2) n equal 2, so n == rank and this matrix invertable

B rank is 2, but this matrix is not square, so this matrix has only left inverse matrix

C rank is 1, this matrix decreases dimensions but does not collapse to linear space because the determinant is more than zero, this matrix has inverse but only left

```python
a = np.linspace(0, 2 * np.pi, 10)
V = np.array([np.cos(a), np.sin(a)]).T

for v in V:
    plt.scatter(*v, c='r')
    x,y = A.dot(np.array(v))
    plt.scatter(x,y, c='b')
    v1 = invA.dot([x,y])
    plt.scatter(*v1, c='green', alpha=0.4)
```

```python
a = np.linspace(0, 2 * np.pi, 10)
V = np.array([np.cos(a), np.sin(a), a * np.pi]).T
for v in V:
    plt.scatter(*v, c='r')
    x,y = B.dot(np.array(v))
    plt.scatter(x,y, c='b')
    v1 = invB.dot([x,y])
    plt.scatter(*v1, c='green', alpha=0.4)
```

```python
fig = plt.figure()
sb = fig.add_subplot(111, projection='3d')

a = np.linspace(0, 2 * np.pi, 10)
V = np.array([np.cos(a), np.sin(a), a * np.pi]).T

for v in V:
    sb.scatter(*v, c='r')
    x,y,z = C.dot(v)
    sb.scatter(x,y,z, c='b')
    sb.plot([v[0],x],[v[1],y],[v[2],z], c='b')
    v1 = invC.dot([x,y,z])
    sb.scatter(*v1, c='green')
    sb.plot([x,v1[0]],[y,v1[1]],[z,v1[2]], c='green')

sb.view_init(110, 50)
```
