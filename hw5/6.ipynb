The triangle corners are represented by the following vectors:

[ (1, 1, 1), (2, 3, -1), (-1, 0, 4) ]

How the area of the triangle changes if we linearly transform its vectors by
the matrix ?

A = [[1 4 0][2 3 0][0 1 2]]

```python
A = np.array([[1,4,0],[2,3,0],[0,1,2]]) # A(n*m) => A(3*3) 
d1 = np.linalg.det(A) # -10.000000000000002
np.linalg.matrix_rank(A) # rank = 3, A(n) == rank, so matirx is not a singularity

a = np.array([1,1,1])
b = np.array([2,3,-1])
c = np.array([-1,0,4])
ab = b-a
ac = c-a
area = 1/2 * np.linalg.norm(np.cross(ab,ac)) # 2.5495097567963922

a1 = A.dot(a)
b1 = A.dot(b)
c1 = A.dot(c)
ab1 = b1-a1
ac1 = c1-a1
area1 = 1/2 * np.linalg.norm(np.cross(ab1,ac1)) # 22.304708023195463

diff = area1 - area # 19.75519826639907
diff2 = area1 / area # 8.748626265789479


a1,b1,c1,d1,area,area1,diff,diff2
```
